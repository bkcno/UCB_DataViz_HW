### Hours and Stations (utilizing data from 2017-2018):

```python
import glob
import pandas as pd

#Most Recent Data from Nov2017 to Feb2018 - Winter months
files = glob.glob("winter_data/*.csv") 

header_saved = False
with open('merged_winter.csv','w') as f_out:
    for filename in files:
        with open(filename) as f_in:
            header = next(f_in)
            if not header_saved:
                f_out.write(header)
                header_saved = True
            for line in f_in:
                f_out.write(line)
```

```python
df = pd.read_csv('merged_winter.csv', error_bad_lines=False, low_memory=False)
df.head()
```

```python
df_start = df.groupby(['start station id', 
                       'start station name', 
                       'start station latitude', 
                       'start station longitude']).size().sort_values(ascending=False).head(30)
popular_starts = df_start.reset_index().rename(columns={'start station id':'station_id','start station name' : 'station_name',
                                              'start station latitude': 'station_lat', 'start station longitude': 'station_lon', 0:'counts' })

popular_starts.to_csv("popular_starts_win.csv")
print("CHECKPOINT: The Top 30 Popular Starting Stations in Winter Are:")
print('_____________________________________________________')
popular_starts.head()
```

```python
df_end = df.groupby(['end station id', 
                       'end station name', 
                       'end station latitude', 
                       'end station longitude']).size().sort_values(ascending=False).head(30)
popular_ends = df_end.reset_index().rename(columns={'end station id':'station_id','end station name' : 'station_name',
                                              'end station latitude': 'station_lat', 'end station longitude': 'station_lon', 0:'counts' })

popular_ends.to_csv("popular_ends_win.csv")
print("CHECKPOINT: The Top 30 Popular Ending Stations in Winter Are:")
print('_____________________________________________________________')
popular_ends.head()
```

```python
df_start_rev = df.groupby(['start station id', 
                       'start station name', 
                       'start station latitude', 
                       'start station longitude']).size().sort_values(ascending=True).head(30)
unpopular_starts = df_start_rev.reset_index().rename(columns={'start station id':'station_id','start station name' : 'station_name',
                                              'start station latitude': 'station_lat', 'start station longitude': 'station_lon', 0:'counts' })

unpopular_starts.to_csv("unpopular_starts_win.csv")
print("CHECKPOINT: The 30 Least Popular Starting Stations in Winter Are:")
print('_________________________________________________________________')
unpopular_starts.head()
```

```python
df_end_rev = df.groupby(['end station id', 
                       'end station name', 
                       'end station latitude', 
                       'end station longitude']).size().sort_values(ascending=True).head(30)
unpopular_ends = df_end_rev.reset_index().rename(columns={'end station id':'station_id','end station name' : 'station_name',
                                              'end station latitude': 'station_lat', 'end station longitude': 'station_lon', 0:'counts' })

unpopular_ends.to_csv("unpopular_ends_win.csv")
print("CHECKPOINT: The 30 Least Popular Ending Stations in Winter are:")
print('_______________________________________________________________')
unpopular_ends.head()
```

```python
popular_bike = df.groupby(['bikeid'])['tripduration'].sum().sort_values(ascending=False)
df_bikes = popular_bike.reset_index().rename(columns={"bikeid":'bike_id', 'tripduration': "total_trips(in hours)"})

df_bikes['total_trips(in hours)'] = pd.Series(["{0:.2f}".format(val / 3600) for val in df_bikes['total_trips(in hours)']], index = df_bikes.index)
df_bikes.head()
```

```python
import glob
#Most recent summer data from Jun2017 to Oct2017
morefiles = glob.glob("summer_data/*.csv") 

header_saved = False
with open('merged_summer.csv','w') as f_out:
    for filename in morefiles:
        with open(filename) as f_in:
            header = next(f_in)
            if not header_saved:
                f_out.write(header)
                header_saved = True
            for line in f_in:
                f_out.write(line)
```

```python
import pandas as pd
df2 = pd.read_csv('merged_summer.csv', error_bad_lines=False, low_memory=False)
df2.head()
```

```python
df2_end = df2.groupby(['end station id', 
                       'end station name', 
                       'end station latitude', 
                       'end station longitude']).size().sort_values(ascending=False).head(30)
popular_ends2 = df2_end.reset_index().rename(columns={'end station id':'station_id','end station name' : 'station_name',
                                              'end station latitude': 'station_lat', 'end station longitude': 'station_lon', 0:'counts' })

popular_ends2.to_csv("popular_ends_sum.csv")
print("CHECKPOINT: The Top 30 Popular Ending Stations in Summer are:")
print('_____________________________________________________________')
popular_ends2.head()
```
